```python
# sets
# no dups, no index(order)

x = {"ben","emil", "mountan-man", "bagdroski","obi-wagwan"}
y = {"caden","jeffrey","jar-jar", "obi-wagwan","mountan-man", "bagdroski" }
overlaps = x.intersection(y)

print(overlaps)

```

```python
# tuple
# non-mutable, dups, iterable, indexed

tuple1 = (1,2,3,4)
tuple_declered = tuple((3,4,5,6))

tuple1.index(4) # returns 3, because 4 is on the indexed 3rd place
tuple1[0] # returns 1

```

```python
# lists

list1 = [1,2,2,3,]
list_declared  = list((1,2,2,3,4))

# index

print(f"first item is always inclusive{list1[1:]}")
print(f"last item is always exlusive {list1[:3]}")
```

```python
# list methods 

list1.append(9) #adding to list
print(list1)
removed = list1.pop() # taking away from list
print(list1)
# change index
list1[1] = 3
# combine list
list1.extend(list_declared)
combined_list1 = [list1, list_declared] # combined as two lists, only two items




```
