```python
sentence = "this is a python course"

words:List[str] = sentence.split(" ")
used_vowels = []

for word in words:
    vowels = ["a", "e", "i", "o", "u", "y"]
    for letter in word:
        if letter in vowels:
            used_vowels.append(letter)

print(used_vowels)
        
    
```

```python
# wehat are funcs

def shorten_names(names_list: List[str], chars: int = 3)-> List[str]:
    """
    Returns a list of strings that have the first 3 carecters of the word with a . appended. 
    
    Args:
        names_list(List[str]): A list of strings that will be shortened
        chars(int): The number of charecters to shorten each list to. 
        Must be greater than 0. Defults to 3
    Returns:
         List[str]: A list of shortend strings to 3 carecters. 
    """
    
    if type(chars)!= int:
        raise TypeError("Chars must be int")
    
    if chars < 0:
        raise ValueError("Chars must be greater than 0")
    names_list = [f"{name[:chars]}." for name in names_list]
    return names_list
    
```

```python
shorten_names(["emil", "ben", "nate", "bomba"])

shorten_names(bar_names, 2)
```
