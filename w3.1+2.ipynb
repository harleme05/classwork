```python
# nested if statments
# type hinting

variable: str = input(f"please enter a name: ")

if variable.isnumeric():
    try:
        num = float(variable)
        if num > 10:
            print("number is to high, start over")
        else:
            print(f"{num} is goooood")
    except Exception as e:
        print(f"error: {e}")
elif variable.isalpha():
    print(f"you chose, {variable.upper()}")
else:
    print("please enter either a number or a letter")


```
