```python
# DF= collection of Series objects

# dict creation
data = {
    "name":["emil", "jakob", "thomas", "Elin"],
    "age":[20,18,55,51],
    "work":["collage", "high school", "surgeon", "project leader"]
}
fam_df = pd.DataFrame(data=data)

fam_df

# Row based creation
import random

names = ["håland", "bolasie", "ødegård", "nusa"]
goals_pg=[random.random() for _ in names ]
games = [round(random.random()*1000) for _ in names]

cols = ["name", "goals/game", "games"]
stats = [(names[i], goals_pg[i], games[i]) for i in range(4)]
stats_df= pd.DataFrame(stats, columns=cols)
stats_df
```

```python
stats_df["games_won"]=[round(random.random()*1000) for _ in names]
stats_df["make_sense?"] = stats_df["games"] > stats_df["games_won"]
stats_df
```

```python
stats_df=stats_df.rename(columns={"goals/game":"goals_per_game"})
stats_df
stats_df.drop(columns="goals_per_game")
```
